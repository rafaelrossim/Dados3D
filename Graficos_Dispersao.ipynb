Gráfico de dispersão

```python
#importando libs
import matplotlib.pyplot as plt
```

```python
#definindo funções
def funcao_ter_grau(x, a=1, b=1, c=1, d=1):
    """uma função que reecbe cinco parâmtros, sendo que quatro últimos são opcionais com default igual a 1"""
    y = a*x**3 + b*x**2 + c*x + d
    return y

def funcao_seg_grau(x, a=1, b=1, c=1):
    y = a*x**2 + b*x + c
    return y
```

```python
#declaração de variáveis
lista_x = range(-10, 10)
lista_y_2 = []
lista_y_3 = []
lista_map = list(map(funcao_seg_grau, lista_x)) #utilizando a funcao map e tranformando o resultado em uma lista.
```

```python
#incrementando nas variáveis
for i in lista_x:
    lista_y_2.append(funcao_ter_grau(i))
    lista_y_3.append(funcao_seg_grau(i, b=-3))
```

Motrando gráficos

```python
#apenas um gráfico
plt.plot(lista_x, lista_y_2, label='Plot', color='#3E7C59')

plt.xlabel("Valores de x")
plt.ylabel("Valores de y")
plt.title("Gráfico de dispersão")
plt.legend()

#salvando o grafico como imagem
plt.savefig(r"D:\Projetos\Python\Dados_3D\imgs\dispersao_1.png", dpi = 300)
```

```python
#dois gráficos
plt.plot(lista_x, lista_y_3, label='Plot', color='#317A86')
plt.scatter(lista_x, lista_y_3, label='Scatter')

plt.xlabel("Valores de x")
plt.ylabel("Valores de y")
plt.title("Gráfico de dispersão")
plt.legend()

#salvando o grafico como imagem
plt.savefig(r"D:\Projetos\Python\Dados_3D\imgs\dispersao_2.png", dpi = 300)
```

```python
#utilizando a funcao map
plt.plot(lista_x, lista_map, label="Map", color='#3E6F6B')
plt.legend()

#salvando o grafico como imagem
plt.savefig(r"D:\Projetos\Python\Dados_3D\imgs\dispersao_map.png", dpi = 300)
```
