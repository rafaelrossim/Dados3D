Gráficos de Histograma

```python
#import libs
import matplotlib.pyplot as plt
from numpy import*
```

```python
#declarando variáveis
n = 10000000 #número de dados gerados
n2 = 5000000
n_col = 50 #número de colunas do histograma

colors = ['#265C4B', '#146551', '#007566', '#589A8D', '#8FC1B5']
colorsl = ['Verde1', 'Verde2', 'Verde3', 'Verde4', 'Verde5']

x = random.randn(n)
y = random.randn(n2)
z = random.randn(n, 5) #esta variável tem 50000000 de registros (n*5)
```

Mostrando Gráficos

```python
#apenas um histograma
plt.hist(x, bins=n_col, density=False, histtype='bar', color='#007566', label='Histograma X')
#bins(int) = quantidade de colunas do histograma
#desity(True) = mostra o densidade de ocorrencias do evento, ex:0.4
#desity(False) = mostra o número de ocorrencias do evento, ex:40
#histtype(string) = tipo de histograma que por default é "bar"

plt.legend()
plt.ylabel("Número de ocorrências")
plt.title("Histograma")

#salvando o grafico como imagem
plt.savefig(r"D:\Projetos\Python\Dados3D\imgs\histograma_1.png", dpi = 300)
```

```python
#gráficos sobrepostos
plt.hist(x, bins=n_col, density=False, histtype='bar', color='#007566', label='Histograma X')
plt.hist(y, bins=n_col, density=False, histtype='bar', color='#589A8D', label='Histograma y')

plt.legend()
plt.ylabel("Número de ocorrências")
plt.title("Histograma")

#salvando o grafico como imagem
plt.savefig(r"D:\Projetos\Python\Dados3D\imgs\histograma_2.png", dpi = 300)
```

```python
#gráficos não sobrepostos
plt.hist(z, bins=25, density=False, histtype='bar', color=colors, label=colorsl)

plt.legend()
plt.ylabel('Número de ocorrências')
plt.title("Histograma")

#salvando o grafico como imagem
plt.savefig(r"D:\Projetos\Python\Dados3D\imgs\histograma_3.png", dpi = 300)
```
