Gráfico de barra

```python
#import libs
import matplotlib.pyplot as plt
from numpy import*
```

```python
#declaração de variáveis
alunos = ['Aluno1', 'Aluno2', 'Aluno3', 'Aluno4', 'Aluno5']
notas_p1 = [20, 25, 17, 10, 22]
notas_p2 = [5, 12, 29, 34, 29]
notas_p3 = [18, 14, 22, 31, 12]

dados_zip = zip(alunos, notas_p1, notas_p2, notas_p3) #aqui contém uma lista de tuplas, ex: (('aluno1', 20, 5, 18), (Aluno2, 25, 12, 14)...)
dados_dict = {dados[0]: max(dados[1], dados[2], dados[3]) for dados in dados_zip} #seleciona somente as maiores notas de cada aluno
#print(dados_dict)

altura = 0.3
y = arange(len(alunos)) #localização das legendas no eixo x
```

```python
#definindo função
def legenda_automatica(grafico):
    """adciona a legenda acima de cada coluna no gráfico, respeitando a altura"""
    for i in grafico:
        comprimento = i.get_width() #descobrindo o comprimento de cada coluna
        plt.annotate(f'{comprimento}', xy=(comprimento, i.get_y() + i.get_height()/2),
        xytext=(9, -3), textcoords='offset points', ha='center')
```

Mostrando gráficos

```python
grafico1 = plt.barh(y - 3*altura/2, notas_p1, height=altura, label="Notas P1", color='#007566', align='edge')
grafico2 = plt.barh(y, notas_p2, height=altura, label="Notas P2", color='#589A8D')
grafico3 = plt.barh(y + altura/2, notas_p3, height=altura, label="Notas P3", color='#146551', align='edge')

plt.ylabel('Pontos')
plt.title('Notas por aluno')
plt.yticks(y, alunos) #usando a localização das legendas no eixo y

#chamando a funcao de legenda automática
legenda_automatica(grafico1)
legenda_automatica(grafico2)
legenda_automatica(grafico3)

plt.legend()

#salvando o gráfico como imagem
plt.savefig(r"D:\Projetos\Python\Dados3D\imgs\barra_1.png", dpi = 300)

plt.show()
```

```python
#utilizando dicionário e funcao zip()
grafico1 = plt.barh(y, dados_dict.values(), height=altura, label="Maior Nota", align="center", color='#007566')

plt.ylabel('Notas')
plt.title('Notas máximas por aluno')
plt.yticks(y, alunos)

#chamando a funcao de legenda automática
legenda_automatica(grafico1)

plt.legend()

#salvando o gráfico como imagem
plt.savefig(r"D:\Projetos\Python\Dados3D\imgs\barra_2.png", dpi = 300)

plt.show()
```
