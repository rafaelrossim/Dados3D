Gráficos de Pizza

```python
#import libs
import matplotlib.pyplot as plt
```

```python
#declarando variáveis
sabores = 'Calabresa', 'Quatro Queijos', 'Marguerita', 'Portuguesa'
apostas = {'Rafa':100, 'Bia':120, 'Ana':90, 'Robert':95, 'Dani':80, 'Cuca':80, 'Luzia':200, 'Gil':250} #supondo que este dicionário veio de uma API
nomes = tuple(apostas.keys())
valores = list(apostas.values())
pedacos = [2, 2, 2, 2] #total de 8 pedaços
destaque_pizza = (0, 0, 0, 0)
destaque_mega = (0, 0, 0, 0, 0, 0, 0.1, 0.1) # destacando os dois últimos valores
```

Mostrando Gráficos

```python
#sabores de pizzas
plt.pie(pedacos, explode=destaque_pizza, labels=sabores, autopct='%1.2f%%', shadow=True, startangle=90)
#explode = permite destacar uma fatia da pizza (recebe valores inteiros)
#autopct = calcula a porcentagem de cada fatia automaticamente (recebe tuplas)

plt.legend(sabores, title='Sabores', loc='right') #loc = localização da legenda
plt.axis("equal") #permite que a caixa de legenda fique responsiva
plt.title("Pizzas")

#salvando o grafico como imagem
plt.savefig(r"D:\Projetos\Python\Dados3D\imgs\pizza.png", dpi = 300)
```

```python
#aposta mega da virada
plt.pie(valores, explode=destaque_mega, labels=nomes, autopct='%1.2f%%', shadow=False, startangle=90)
plt.legend(nomes, title='Apostadores', loc='best')
plt.axis("equal")
plt.title("Mega da Virada")

#salvando o grafico como imagem
plt.savefig(r"D:\Projetos\Python\Dados3D\imgs\pizza_apostadores.png", dpi = 300)
```
